median, or mode) or drop them. Justify your action for this task.

```python

df = df.drop('New_Price', axis=1)
for column in df.columns:
  if df[column].isnull().any():
    print(f"Column '{column}' has missing values.")
    if pd.api.types.is_numeric_dtype(df[column]):
      df[column].fillna(df[column].median(), inplace=True)
      print(f"Imputed missing values in '{column}' with the median.")
    else:
      df[column].fillna(df[column].mode()[0], inplace=True)
      print(f"Imputed missing values in '{column}' with the mode.")
print("\nMissing values after imputation/removal:")
print(df.isnull().sum())
```

b.Remove the units from some of the attributes and only keep the numerical values (for
example remove kmpl from “Mileage”, CC from “Engine”, bhp from “Power”, and lakh from
“New_price”).

```python
# Removing units from 'Mileage', 'Engine', 'Power' varaibles
df11['Mileage'] = df11['Mileage'].str.extract('(\d+)').astype(float)
df11['Engine'] = df11['Engine'].str.extract('(\d+)').astype(float)
df11['Power'] = df11['Power'].str.extract('(\d+)').astype(float)

# Removing units from 'New_Price' varaible and converting to numerical
df11['New_Price'] = df11['New_Price'].str.extract('(\d+)').astype(float)


# checking any missing values are there in data
missing_values = df11.isnull().sum()
print(missing_values)
df11.to_csv('removing_units.csv', index=False)
```
